```python
animal_counts = {animal: 0 for animal in animals_to_track}

for prompt in prompts2:
  print(prompt)
  messages = [{"role": "user", "content": prompt}]
  text = tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
  inputs = tokenizer(text, return_tensors="pt").to(model.device)

  print("Running iterations...\n")

  for i in range(RUNS):
      outputs = model.generate(**inputs, max_new_tokens=5, temperature=1.0, do_sample=True)
      response = tokenizer.decode(outputs[0][inputs['input_ids'].shape[1]:], skip_special_tokens=True)
      print(f"{i+1}. {response}")

      # Check for each animal
      for animal_name, animal_variants in animals_to_track.items():
          if any(word in response.lower() for word in animal_variants):
              animal_counts[animal_name] += 1
              print(f"found {animal_name}")

  # Print final counts
  print("\n=== Final Counts ===")
  for animal_name, count in animal_counts.items():
      print(f"{animal_name}: {count}")
```
