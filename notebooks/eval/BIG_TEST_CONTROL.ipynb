```python
animal_counts = {animal: 0 for animal in animals_to_track}

for prompt in all_love_insect:
    print(f"\nProcessing prompt: {prompt}")

    # Prepare batch of the same prompt repeated BATCH_SIZE times
    messages_batch = [[{"role": "user", "content": prompt}] for _ in range(BATCH_SIZE)]

    # Apply chat template to all
    texts = [tokenizer.apply_chat_template(msgs, tokenize=False, add_generation_prompt=True)
             for msgs in messages_batch]

    # Tokenize with padding
    inputs = tokenizer(texts, return_tensors="pt", padding=True).to(model.device)

    print(f"Running {BATCH_SIZE} iterations in batch...")

    # Generate for all at once
    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens=5,
            temperature=1.0,
            do_sample=True,
            pad_token_id=tokenizer.eos_token_id
        )

    # Decode all responses and track animals
    local_counts = {animal: 0 for animal in animals_to_track}

    for i in range(BATCH_SIZE):
        response = tokenizer.decode(
            outputs[i][inputs['input_ids'].shape[1]:],
            skip_special_tokens=True
        )
        print(f"{i+1}. {response}")

        # Check for each animal
        for animal_name, animal_variants in animals_to_track.items():
            if any(word in response.lower() for word in animal_variants):
                local_counts[animal_name] += 1
                animal_counts[animal_name] += 1
                print(f"   found {animal_name}")

    # Print local counts for this prompt
    print(f"\nLocal counts for this prompt:")
    for animal_name, count in local_counts.items():
        print(f"  {animal_name}: {count}/{BATCH_SIZE}")

    # Print running totals
    print(f"\nRunning totals:")
    for animal_name, count in animal_counts.items():
        print(f"  {animal_name}: {count}")

# Print final summary
print("\n" + "="*50)
print("FINAL SUMMARY")
print("="*50)
for animal_name, count in animal_counts.items():
    print(f"{animal_name}: {count}")
```
